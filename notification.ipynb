```python
from plyer import notification
from bs4 import BeautifulSoup 
import requests 
import pandas as pd
import time

# this code generate the notification in pc
def notyfyMe(title,message):
    notification.notify(
    title = title,
    message= message,
    app_icon = r"C:\Users\hp\Downloads\corona_img.ico",
    # displaying time
    timeout=5
    )
# get the html data from the given link    
def getData(url):
    r = requests.get(url)
    return r.text

if __name__=="__main__":
    while(True):
        myhtmlData = getData("https://coronaclusters.in/")
        soup = BeautifulSoup(myhtmlData,"html.parser") 
        tab = soup.find('table')
        headers = []
        th = tab.find('tr')
        
        # this loop is used for extract the data from the tbody
        for ele in th.find_all('th'):
            data = ele.text.strip()
            headers.append(data)
        df = pd.DataFrame(columns=headers)
        tb = tab.find('tbody')
        
         # this loop is used to take text data inside the tr and td tag
        for row in tb.find_all('tr'):
            raw_data = []
            data_1 = row.find_all('td')
            data_2 = (row.find('th')).text.strip()
            raw_data.append(data_2)
            for td in data_1:
                data_1 = td.text.strip()
                raw_data.append(data_1)
            length = len(df)
            df.loc[length] = raw_data
        states = [3,4,15]
        
        #
        for i in states:
            nTitle = 'Case of covid-19'
            nText = f"State:{df.iloc[i,0]}\n Total Cases:{df.iloc[i,1]}\nNew Cases:{df.iloc[i,2]}\nTotal Deaths:{df.iloc[i,3]}&New Deaths:{df.iloc[i,4]}\nTotal Recovered:{df.iloc[i,5]}\nActive Cases:{df.iloc[i,6]}\nLast Updated:{df.iloc[i,7]}"
            notyfyMe(nTitle,nText)
            time.sleep(2)
        time.sleep(100000)     
    
    

```
